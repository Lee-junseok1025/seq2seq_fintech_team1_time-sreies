```python
import os
import numpy as np
import pandas as pd

import pyupbit as py
```

```python
py.get_tickers(fiat='USD')
```

```python
py.get_current_price(['USDT-BTC','USDT-ETH'])
```

```python
tickers = ['USDT-BTC','USDT-ETH','USDT-XRP','USDT-ADA','USDT-LTC']
interval = 'minute60'
```

```python
from tqdm import tqdm
```

```python
coin_set = []
for ticker in tqdm(tickers):
    coin = py.get_ohlcv(ticker=ticker,count=20000,interval=interval,to='2022-05-01')
    coin_set.append(coin)
```

```python
BTC = coin_set[0]
ETH = coin_set[1]
XRP = coin_set[2]
ADA = coin_set[3]
LTC = coin_set[4]
```

```python
BTC.to_csv('BTC.csv',index=True)
ETH.to_csv('ETH.csv')
XRP.to_csv('XRP.csv')
ADA.to_csv('ADA.csv')
LTC.to_csv('LTC.csv')
```
